```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

data_2022 = {
    'Country': ['China', 'UK', 'India', 'South Korea', 'Russian Federation', 'Brazil', 'Hong Kong (SAR China)', 'Mexico', 'South Africa', 'Japan', 'Vietnam', 'Nigeria', 'Indonesia', 'Argentina',
                'Egypt', 'Saudi Arabia', 'Colombia', 'Chile', 'Kenya', 'Ghana'],
    'Actual Net HNWI Outflows 2022': [-10800, -7500, -1600, -400, 8500, -1800, -2400, -800, -400, -100, -200, -200, -400, -200, -200, -600, -100, -100, -100, -100]
}

df_2022 = pd.DataFrame(data_2022)
print("2022 Dataset:")
print(df_2022)
```

```python


data_2023 = {
    'Country': ['China', 'India', 'UK', 'Russian Federation', 'Brazil', 'Hong Kong (SAR China)', 'South Korea', 'Mexico', 'South Africa', 'Japan', 'Vietnam', 'Nigeria', 'Indonesia', 'Argentina',
                'Egypt', 'Saudi Arabia', 'Colombia', 'Chile', 'Kenya', 'Ghana'],
    'Forecast Net HNWI Outflows 2023': [-13500, -6500, -3200, -3000, -1200, -1000, -800, -700, -500, -300, -300, -300, -200, -200, -200, -100, -100, -100, -100, -100]
}

df_2023 = pd.DataFrame(data_2023)
print("\n2023 Dataset:")
print(df_2023)
```

```python


data_2024 = {
    'Country': ['China', 'UK', 'India', 'South Korea', 'Russian Federation', 'Brazil', 'South Africa', 'Taiwan (Chinese Taipei)', 'Nigeria', 'Vietnam'],
    'Projected net outflows of millionaires in 2024': [-15200, -9500, -4300, -1200, -1000, -800, -600, -400, -300, -300]
}

df_2024 = pd.DataFrame(data_2024)
print("\n2024 Dataset:")
print(df_2024)
```

```python


# Merging the DataFrames
merged_df = df_2022.merge(df_2023, on='Country').merge(df_2024, on='Country')

# Renaming the columns
merged_df.columns = ['Country', 'Actual Net HNWI Outflows 2022', 'Forecast Net HNWI Outflows 2023', 'Projected Net Outflows 2024']

print("\nComparative Analysis Across Years:")
print(merged_df)
```

```python

# Plotting
plt.style.use('seaborn-darkgrid')
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

axes[0].bar(merged_df['Country'], merged_df['Actual Net HNWI Outflows 2022'], color='blue')
axes[0].set_title('Net HNWI Outflows in 2022')
axes[0].set_ylabel('Net HNWI Outflows')
axes[0].tick_params(axis='x', rotation=90)

axes[1].bar(merged_df['Country'], merged_df['Forecast Net HNWI Outflows 2023'], color='green')
axes[1].set_title('Net HNWI Outflows in 2023')
axes[1].tick_params(axis='x', rotation=90)

axes[2].bar(merged_df['Country'], merged_df['Projected Net Outflows 2024'], color='orange')
axes[2].set_title('Net HNWI Outflows in 2024')
axes[2].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
